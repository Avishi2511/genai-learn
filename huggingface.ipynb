```python
from langchain import PromptTemplate, HuggingFaceHub, LLMChain
```

```python
import os
os.environ['HuggingFaceHub_API_TOKEN'] = ''
```

```python
prompt = PromptTemplate(
    input_variables=["input"],
    template="What is a good name for a company that makes {input}?",
)
```

```python
facebook/mbart-large-50
```

```python
chain = LLMChain(llm=HuggingFaceHub(repo_id = 'facebook/mbart-large-50', model_kwargs={"temperature": 1.5,}), prompt=prompt)
```

```python
chain.run('colorful clothes')
```

### Text Generation Models | Decoder Only Models

```python
from langchain import HuggingFacePipeline
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, AutoModelForSeq2SeqLM
```

```python
prompt = PromptTemplate(
    input_variables=["name"],
    template="Can you tell me abt footballer {name}?",
)
```

```python
model_id = 'facebook/mbart-large-50'
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained(model_id, device_map="auto")  
```

```python
pipeline = pipeline("text2text-generation", model=model, tokenizer=tokenizer, device_map="auto")
```

```python
local_llm = HuggingFacePipeline(pipeline=pipeline)
```

```python
chain = LLMChain(llm=local_llm, prompt=prompt)
```

```python
chain.run("messi")
```
